    The E-Commerce company has the following rule for computing their own revenue: </font> 
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

```python
for rev, row in ECom_Data.iterrows():
    if row['discounted_price']>600:
        ECom_Data.loc[rev,'company revenue']=row['discounted_price']*0.25
    elif row['discounted_price']<=600 and row['discounted_price']>350:
        ECom_Data.loc[rev,'company revenue']=row['discounted_price']*0.15
    elif row['discounted_price']<=350 and row['discounted_price']>100:
        ECom_Data.loc[rev,'company revenue']=row['discounted_price']*0.10
    else:
        ECom_Data.loc[rev,'company revenue']=row['discounted_price']*0.05   
```

```python
ECom_Data
```

```python
print("Net revenue generated by the E-Commerce company over all orders placed is",ECom_Data['company revenue'].sum())
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and 
## iv.	total net BrandRevenue generated per Brand. 

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
w=pd.DataFrame(ECom_Data.groupby(['Brand'])['Customer_uniq_id'].nunique()) #new dataframe for oder placed per brand
w
```

```python
x=pd.DataFrame(ECom_Data.groupby(['Brand'])['retail_price'].sum()) #new drataframe for retail price per band
x
```

```python
y=pd.DataFrame(ECom_Data.groupby(['Brand'])['discounted_price'].sum()) #new drataframe for discount price per band
y
```

```python
z=pd.DataFrame(ECom_Data.groupby(['Brand'])['Brand revenue'].sum()) #new drataframe for retail price per band
z
```

```python
w_x=pd.merge(w,x,how='outer',on='Brand') #merge w and x
w_x_y=pd.merge(w_x,y,how='outer',on='Brand')
w_x_y_z=pd.merge(w_x_y,z,how='outer',on='Brand')
w_x_y_z
```

```python
w_x_y_z.rename(columns={'Customer_uniq_id':'Total order placed','retail_price':'Total Retail Price','discounted_price':'Total_discounted_price','Brand revenue':'Total_brand_revenue'},inplace=True)
#rename columns
w_x_y_z
```

```python
plt.figure(figsize=(10,5))
sns.pairplot(data=w_x_y_z[['Total order placed','Total Retail Price','Total_discounted_price','Total_brand_revenue']]);
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly net Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

```python
ECom_Data
```

```python
ECom_Data['Order_Date']  = pd.to_datetime(ECom_Data['Order_Date'],dayfirst=True)
ECom_Data.info()
```

```python
ECom_Data['month'] = ECom_Data['Order_Date'].dt.month
ECom_Data
```

```python
ECom_Data['year'] = ECom_Data['Order_Date'].dt.year
ECom_Data
```

```python
(ECom_Data.groupby(['year'])['month'].unique())#year 2020 contains order from december month only.
```

```python
ECom_Data_2021=ECom_Data.loc[ECom_Data['year']!=2020]
ECom_Data_2021 #we are considering orders placed only in the year 2021, all orders placed in 2020 is in december month only. we are creating line plot for the year 2021 only.
```

```python
ECom_Data_north=ECom_Data_2021.loc[ECom_Data_2021['Region']=='North']
ECom_Data_north
```

```python
ECom_Data_north_monthy_net_revenue=pd.DataFrame(ECom_Data_north.groupby(['month'])['company revenue'].sum())
ECom_Data_north_monthy_net_revenue#created a dataframe for monthly net revenue for north region
```

```python
fig = plt.figure(figsize = (20,7))
plt.title('LIneplot:Monthly net revenue - North region')
sns.lineplot(data = ECom_Data_north_monthy_net_revenue , x = 'month' , y = 'company revenue',ci=False)
fig.autofmt_xdate()
plt.ylabel('Revenue')
```
